```python
%load_ext autoreload
%autoreload 2

import sys
import featuregraph as fg

from featuregraph.accumulation.accumulations import add_accumulation_signal, add_accumulation_id, add_accumulation_primitives, add_accumulation_features, get_accumulation_summary

print("Python:", sys.executable)
print("FeatureGraph:", fg.__file__)
```

```python
bidmc = fg.datasets.bidmc(
    subject=1,
)
```

```python
bidmc = fg.oscillation.waves.add_wave_primitives(bidmc, ['respiration'])
bidmc = fg.oscillation.waves.add_wave_id(bidmc, ['respiration'], 'subject')
bidmc = fg.oscillation.waves.add_wave_features(bidmc, ['respiration'], 'respiration_wave_id')
bidmc
```

```python
bidmc = add_accumulation_signal(
    bidmc,
    signals=["respiration"],
    thresholds={'respiration': 'group_mean'},
    group='respiration_wave_id'
)

bidmc = add_accumulation_id(
    bidmc,
    signals=["respiration"],
)

bidmc = add_accumulation_primitives(
    bidmc,
    signals=["respiration"],
)

bidmc = add_accumulation_features(
    bidmc,
    signals=["respiration"],
)

summarydf = get_accumulation_summary(
    bidmc,
    signal="respiration",
)
```
